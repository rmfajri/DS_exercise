#### Importing the libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Read the iris dataset

```python
#dataset = pd.read_csv('Dataset/Mall_Customers.csv')
dataset = pd.read_csv('Dataset/iris.csv')
X=dataset.values
```

#### Using the dendogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show() 
```

#### Fitting hierarchical clustering to the iris dataset 3 cluster

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

#### Visualizing the cluster the class=setosa,versicolor,virginica

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'setosa')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'versicolor')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'virginica')

plt.title('Clusters of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```
